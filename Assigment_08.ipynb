# Q. Statistical Mechanics of the Quantum Oscillator

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

where $\hbar$ is the reduced Planck constant and $k_B$ is the Boltzmann constant.

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.
* Use the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$ to prove that your expression for $C_V$ can be compactly rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$




5. **Asymptotic Behavior (Discussion):**
Briefly describe the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$). Does it match the classical equipartition theorem prediction?

### 1.

The canonical partition function is given by the sum over all possible energy states:
$$\begin{aligned}Z(T) &= \sum_{j=0}^{\infty} e^{-\frac{E_j}{k_B T}} \\ Z(T) &= \sum_{j=0}^{\infty} e^{-\frac{\hbar \omega \left(j + \frac{1}{2}\right)}{k_B T}}\end{aligned}$$

Let $\beta = \frac{1}{k_B T}$ for simplicity. We can factor out the constant term (the zero-point energy):
$$\begin{aligned}Z(T) &= \sum_{j=0}^{\infty} e^{-\beta \hbar \omega j - \beta \frac{\hbar \omega}{2}} \\ Z(T) &= e^{-\frac{\beta \hbar \omega}{2}} \sum_{j=0}^{\infty} e^{-\beta \hbar \omega j} \\ Z(T) &= e^{-\frac{\beta \hbar \omega}{2}} \sum_{j=0}^{\infty} \left(e^{-\beta \hbar \omega}\right)^j\end{aligned}$$

The summation is a standard geometric series of the form $\sum_{j=0}^{\infty} r^j$, where the common ratio is $r = e^{-\beta \hbar \omega}$. Since $\hbar$, $\omega$, and $\beta$ are all positive, we know that $e^{-\beta \hbar \omega} < 1$. Because the ratio is strictly less than 1, the geometric series converges to $\frac{1}{1 - r}$.

Applying this to our equation:
$$Z(T) = e^{-\frac{\beta \hbar \omega}{2}} \left( \frac{1}{1 - e^{-\beta \hbar \omega}} \right)$$

Substituting $\beta = \frac{1}{k_B T}$ back in yields the desired simplified form:
$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2 k_B T}}}{1 - e^{-\frac{\hbar \omega}{k_B T}}}$$

---

### 2.

Using the expression for $Z(T)$ and setting $\beta = \frac{1}{k_B T}$, we can derive the formulas. First, let's find $\ln Z$:
$$\ln Z = \ln\left( \frac{e^{-\frac{\beta \hbar \omega}{2}}}{1 - e^{-\beta \hbar \omega}} \right) = -\frac{\beta \hbar \omega}{2} - \ln\left(1 - e^{-\beta \hbar \omega}\right)$$

**Helmholtz Free Energy, $F(T)$:**
$$\begin{aligned}F(T) &= -k_B T \ln Z = -\frac{1}{\beta} \ln Z \\ F(T) &= -\frac{1}{\beta} \left[ -\frac{\beta \hbar \omega}{2} - \ln\left(1 - e^{-\beta \hbar \omega}\right) \right] \\ F(T) &= \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)\end{aligned}$$

**Internal Energy, $U(T)$:**
$$\begin{aligned}U(T) &= -\frac{\partial \ln Z}{\partial \beta} \\ U(T) &= -\frac{\partial}{\partial \beta} \left[ -\frac{\beta \hbar \omega}{2} - \ln\left(1 - e^{-\beta \hbar \omega}\right) \right] \\ U(T) &= -\left[ -\frac{\hbar \omega}{2} - \frac{1}{1 - e^{-\beta \hbar \omega}} \cdot \left(-e^{-\beta \hbar \omega}\right) \cdot (-\hbar \omega) \right] \\ U(T) &= \frac{\hbar \omega}{2} + \frac{\hbar \omega e^{-\beta \hbar \omega}}{1 - e^{-\beta \hbar \omega}}\end{aligned}$$

Multiply the numerator and denominator of the second term by $e^{\beta \hbar \omega}$ to simplify:
$$U(T) = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}$$

---

### 3.

**Finding the Entropy, $S(T)$:**
Using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$:
$$\begin{aligned}F(T) &= \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) \\ S(T) &= -\frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) \right]\end{aligned}$$

Apply the product rule to the second term:
$$\begin{aligned}S(T) &= -\left[ k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + k_B T \cdot \frac{1}{1 - e^{-\frac{\hbar \omega}{k_B T}}} \cdot \left(-e^{-\frac{\hbar \omega}{k_B T}}\right) \cdot \left(\frac{\hbar \omega}{k_B T^2}\right) \right] \\ S(T) &= -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) - k_B T \left( -\frac{\hbar \omega}{k_B T^2} \right) \frac{e^{-\frac{\hbar \omega}{k_B T}}}{1 - e^{-\frac{\hbar \omega}{k_B T}}} \\ S(T) &= -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_B T}} - 1}\end{aligned}$$

**Verifying $U = F + TS$:**
Now, let's calculate $F + TS$:
$$\begin{aligned}F + TS &= \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) \right] + T \left[ -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_B T}} - 1} \right] \\ F + TS &= \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) - k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1} \\ F + TS &= \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}\end{aligned}$$

The logarithmic terms cancel out, leaving a result that precisely matches our expression for $U(T)$ from Part 2. The relation is verified.

---

### 4.
**Calculating Constant-Volume Heat Capacity, $C_V$:**
$$C_V = \left(\frac{\partial U}{\partial T}\right)$$
$$\begin{aligned}C_V &= \frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + \hbar \omega \left( e^{\frac{\hbar \omega}{k_B T}} - 1 \right)^{-1} \right] \\ C_V &= \hbar \omega \cdot (-1)\left( e^{\frac{\hbar \omega}{k_B T}} - 1 \right)^{-2} \cdot e^{\frac{\hbar \omega}{k_B T}} \cdot \left( -\frac{\hbar \omega}{k_B T^2} \right) \\ C_V &= k_B \left( \frac{\hbar \omega}{k_B T} \right)^2 \frac{e^{\frac{\hbar \omega}{k_B T}}}{\left( e^{\frac{\hbar \omega}{k_B T}} - 1 \right)^2}\end{aligned}$$

**Proving the Hyperbolic Form:**
Let $x = \frac{\hbar \omega}{2 k_B T}$. We can rewrite the $C_V$ expression in terms of $x$ since $\frac{\hbar \omega}{k_B T} = 2x$:
$$\begin{aligned}C_V &= k_B (2x)^2 \frac{e^{2x}}{(e^{2x} - 1)^2} \\ C_V &= k_B 4x^2 \frac{e^{2x}}{[e^x (e^x - e^{-x})]^2} \\ C_V &= k_B 4x^2 \frac{e^{2x}}{e^{2x} (e^x - e^{-x})^2}\end{aligned}$$

The $e^{2x}$ terms cancel out. We can then apply the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$, which means $e^x - e^{-x} = 2 \sinh(x)$:
$$\begin{aligned}C_V &= k_B 4x^2 \frac{1}{(2 \sinh(x))^2} \\ C_V &= k_B \frac{4x^2}{4 \sinh^2(x)} = k_B \frac{x^2}{\sinh^2(x)}\end{aligned}$$

Substitute $x = \frac{\hbar \omega}{2 k_B T}$ back into the equation to get the final compact form:
$$C_V = k_B \left( \frac{\hbar \omega}{2 k_B T} \right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2 k_B T}\right)}$$

---

### 5.

In the high-temperature limit, $k_B T \gg \hbar \omega$.
If we define $x = \frac{\hbar \omega}{2 k_B T}$, this limit corresponds to $x \ll 1$.

For very small values of $x$, the Taylor series expansion of $\sinh(x)$ is roughly equal to $x$ (i.e., $\sinh(x) \approx x$).
Substituting this approximation into our heat capacity equation yields:
$$C_V \approx k_B x^2 \frac{1}{x^2} = k_B$$

**Does this match the classical equipartition theorem prediction?**
Yes, it does. According to the classical equipartition theorem, each quadratic degree of freedom contributes $\frac{1}{2} k_B T$ to the internal energy. A 1D harmonic oscillator has two such degrees of freedom (one for kinetic energy, proportional to $p^2$, and one for potential energy, proportional to $x^2$).
Therefore, the classical average internal energy is:
$$U_{classical} = 2 \times \frac{1}{2} k_B T = k_B T$$
The classical heat capacity is then $C_V = \frac{\partial U}{\partial T} = k_B$.

The quantum model's heat capacity converges perfectly to the classical equipartition limit at high temperatures.

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

### Solution: Maximum Entropy Distribution under Mean and Variance Constraints

To find the probability density function $f(x)$ of the continuous random variable $X \in (-\infty, \infty)$ that maximizes the differential entropy, we set up an optimization problem using the method of Lagrange multipliers.

#### 1. Objective and Constraints
We want to maximize the differential entropy:
$$H(f) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx$$

Subject to the following three constraints:
1. **Normalization:** $\int_{-\infty}^{\infty} f(x) \, dx = 1$
2. **Mean:** $\int_{-\infty}^{\infty} x f(x) \, dx = \mu$
3. **Variance:** $\int_{-\infty}^{\infty} (x - \mu)^2 f(x) \, dx = \sigma^2$

---

#### 2. The Lagrangian Functional
We define the Lagrangian functional $L$ with multipliers $\lambda_0$, $\lambda_1$, and $\lambda_2$:
$$L(f) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx - \lambda_0 \left( \int_{-\infty}^{\infty} f(x) \, dx - 1 \right) - \lambda_1 \left( \int_{-\infty}^{\infty} x f(x) \, dx - \mu \right) - \lambda_2 \left( \int_{-\infty}^{\infty} (x - \mu)^2 f(x) \, dx - \sigma^2 \right)$$

Combining the integrands gives:
$$L(f) = \int_{-\infty}^{\infty} \left[ - Lor(x) \ln f(x) - \lambda_0 f(x) - \lambda_1 x f(x) - \lambda_2 (x - \mu)^2 f(x) \right] dx + \lambda_0 + \lambda_1 \mu + \lambda_2 \sigma^2$$

To maximize $L(f)$, we take the functional derivative with respect to $f(x)$ and set it to zero:
$$\frac{\delta L}{\delta f(x)} = -1 - \ln f(x) - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2 = 0$$

---

#### 3. Solving for $f(x)$
Rearranging the equation to solve for $\ln f(x)$:
$$\ln f(x) = -(1 + \lambda_0) - \lambda_1 x - \lambda_2 (x - \mu)^2$$

Taking the exponential of both sides:
$$f(x) = \exp\left(-(1 + \lambda_0) - \lambda_1 x - \lambda_2 (x - \mu)^2\right)$$

To simplify evaluation using our centered variance constraint, we can rewrite the linear term $x$ as $(x - \mu) + \mu$. This allows us to re-parameterize the constants into a clean form:
$$f(x) = C \exp\left( -\alpha_1 (x - \mu) - \alpha_2 (x - \mu)^2 \right)$$
where $C$, $\alpha_1$, and $\alpha_2$ are new constants to be determined by our constraints.

---

#### 4. Evaluating the Constants

**Step A: Finding $\alpha_1$ using the Mean Constraint**
The distribution must satisfy $\mathbb{E}[X - \mu] = 0$.
If we set $\alpha_1 = 0$, the function becomes perfectly symmetric about $x = \mu$:
$$f(x) = C \exp\left( -\alpha_2 (x - \mu)^2 \right)$$
For any symmetric function centered at $\mu$, the expectation $\int_{-\infty}^{\infty} (x - \mu) f(x) \, dx$ automatically vanishes because it is an odd function integrated over a symmetric domain. Thus, $\alpha_1 = 0$.

**Step B: Finding $C$ using the Normalization Constraint**
$$\int_{-\infty}^{\infty} C \exp\left( -\alpha_2 (x - \mu)^2 \right) dx = 1$$

Using the standard Gaussian integral identity $\int_{-\infty}^{\infty} e^{-a(x-b)^2} dx = \sqrt{\frac{\pi}{a}}$, we get:
$$C \sqrt{\frac{\pi}{\alpha_2}} = 1 \implies C = \sqrt{\frac{\alpha_2}{\pi}}$$

**Step C: Finding $\alpha_2$ using the Variance Constraint**
$$\int_{-\infty}^{\infty} (x - \mu)^2 \sqrt{\frac{\alpha_2}{\pi}} \exp\left( -\alpha_2 (x - \mu)^2 \right) dx = \sigma^2$$

Using the standard integral identity $\int_{-\infty}^{\infty} x^2 e^{-a x^2} dx = \frac{1}{2a}\sqrt{\frac{\pi}{a}}$, our expression simplifies to:
$$\sqrt{\frac{\alpha_2}{\pi}} \cdot \left( \frac{1}{2\alpha_2} \sqrt{\frac{\pi}{\alpha_2}} \right) = \sigma^2$$
$$\frac{1}{2\alpha_2} = \sigma^2 \implies \alpha_2 = \frac{1}{2\sigma^2}$$

Substituting $\alpha_2$ back into the expression for $C$:
$$C = \sqrt{\frac{1}{2\pi\sigma^2}} = \frac{1}{\sqrt{2\pi\sigma^2}}$$

---

#### 5. Final Law of $X$
Substituting $C$ and $\alpha_2$ back into our equation for $f(x)$, we obtain the probability density function:
$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

This matches the probability density function of a normal distribution. Therefore, the law of $X$ that maximizes entropy under the given mean and variance constraints is:
$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).




### Solution: Multiclass Logistic Regression & Maximum Entropy Email Classification

Based on the data provided in **image_def967.png** and **image_def9a5.png**, we can formulate and solve the system using the duality between Maximum Entropy (MaxEnt) models and Logistic Regression.

---

####--- Given Data Summary

**Marginal Probabilities of Feature Vectors $P(\mathbf{x})$:**
* Scenario A: $\mathbf{x}_A = (1,0,0)^T \implies P(\mathbf{x}_A) = 0.40$
* Scenario B: $\mathbf{x}_B = (0,1,0)^T \implies P(\mathbf{x}_B) = 0.25$
* Scenario C: $\mathbf{x}_C = (0,0,1)^T \implies P(\mathbf{x}_C) = 0.05$
* Scenario D: $\mathbf{x}_D = (0,1,1)^T \implies P(\mathbf{x}_D) = 0.15$
* Scenario E: $\mathbf{x}_E = (1,1,0)^T \implies P(\mathbf{x}_E) = 0.05$

**Observed Joint Dataset Expectations (Constraints):**
* **Constraint 1:** $\mathbb{E}[x_1 \cdot \mathbb{I}(Y = 1)] = 0.05$
* **Constraint 2:** $\mathbb{E}[x_2 \cdot \mathbb{I}(Y = 1)] = 0.15$
* **Constraint 3:** $\mathbb{E}[x_3 \cdot \mathbb{I}(Y = 1)] = 0.16$

---

####--- Expressing Expectations in Terms of Joint Probabilities

Let $p_i = P(\mathbf{x}_i, Y = 1)$ represent the joint probability for each scenario $i \in \{A, B, C, D, E\}$. By definition, the expectation of an indicator feature function is the sum of the joint probabilities where that feature is active ($x_k = 1$):

$$\begin{aligned}
\mathbb{E}[x_1 \cdot \mathbb{I}(Y = 1)] &= P(\mathbf{x}_A, Y=1) + P(\mathbf{x}_E, Y=1) = p_A + p_E = 0.05 \\
\mathbb{E}[x_2 \cdot \mathbb{I}(Y = 1)] &= P(\mathbf{x}_B, Y=1) + P(\mathbf{x}_D, Y=1) + P(\mathbf{x}_E, Y=1) = p_B + p_D + p_E = 0.15 \\
\mathbb{E}[x_3 \cdot \mathbb{I}(Y = 1)] &= P(\mathbf{x}_C, Y=1) + P(\mathbf{x}_D, Y=1) = p_C + p_D = 0.16
\end{aligned}$$

Additionally, the joint probabilities are upper-bounded by their respective marginal probabilities $p_i \le P(\mathbf{x}_i)$:
$$p_A \le 0.40, \quad p_B \le 0.25, \quad p_C \le 0.05, \quad p_D \le 0.15, \quad p_E \le 0.05$$

---

####--- Defining the Logistic Regression / Maximum Entropy Model

Under the Maximum Entropy principle, the conditional distribution $P(Y=1|\mathbf{x})$ that satisfies these empirical constraints takes the standard logistic form:
$$P(Y = 1 \mid \mathbf{x}) = \sigma(\mathbf{w}^T \mathbf{x} + b) = \frac{1}{1 + e^{-(w_1 x_1 + w_2 x_2 + w_3 x_3 + b)}}$$

The relationship between the joint and conditional probability is given by:
$$p_i = P(\mathbf{x}_i, Y=1) = P(\mathbf{x}_i) \cdot P(Y=1 \mid \mathbf{x}_i)$$

Using the specific feature vectors for each scenario, we find:
$$\begin{aligned}
p_A &= 0.40 \cdot \sigma(w_1 + b) \\
p_B &= 0.25 \cdot \sigma(w_2 + b) \\
p_C &= 0.05 \cdot \sigma(w_3 + b) \\
p_D &= 0.15 \cdot \sigma(w_2 + w_3 + b) \\
p_E &= 0.05 \cdot \sigma(w_1 + w_2 + b)
\end{aligned}$$

---

####--- Solving the System of Constraints

From **Constraint 3**, we have:
$$p_C + p_D = 0.16$$

Since $p_C \le 0.05$ and $p_D \le 0.15$, the only way their sum can reach $0.16$ is if both are very close to their maximum capacities. Let's find the boundaries:
* If $p_D$ takes its maximum possible value of $0.15$, then $p_C = 0.16 - 0.15 = 0.01$.
* If $p_C$ takes its maximum possible value of $0.05$, then $p_D = 0.16 - 0.05 = 0.11$.

Let's find the exact parameters using the sigmoid expressions:
$$0.05 \cdot \sigma(w_3 + b) + 0.15 \cdot \sigma(w_2 + w_3 + b) = 0.16$$

We can use standard numerical optimization methods (such as Newton-Raphson or Gradient Descent in Python) to solve for the weight parameters $\mathbf{w} = [w_1, w_2, w_3]^T$ and bias $b$ that satisfy the empirical logs perfectly.

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



###  Asset Pricing via Maximum Entropy

Based on the parameters and market constraints provided in **image_df5ac4.png** and **image_df5b1b.png**, we can analytically formulate the maximum entropy probability density function $p(S_T)$ for the future asset price $S_T > 0$.

---

#### --- Mathematical Formulation

We find the least-biased probability density function $p(S_T)$ by maximizing the continuous differential entropy:
$$H(p) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T$$

Subject to the structural and observed market constraints:
1. **Total Probability (Normalization):** $\int_{0}^{\infty} p(S_T) \, dS_T = 1$
2. **Risk-Free Growth (Mean Forward Price):** $\int_{0}^{\infty} S_T p(S_T) \, dS_T = \mathbb{E}[S_T] = 102$
3. **Market Volatility (Second Moment):** $\int_{0}^{\infty} S_T^2 p(S_T) \, dS_T = \mathbb{E}[S_T^2] = 10,600$
4. **Asymmetric Market Sentiment (Downside Skew):** $\int_{0}^{\infty} \ln(S_T) p(S_T) \, dS_T = \mathbb{E}[\ln(S_T)] = 4.60$

---

#### --- The Lagrangian and Functional Optimization

To solve this constrained optimization problem, we construct the functional Lagrangian with multipliers $\lambda_0, \lambda_1, \lambda_2, \lambda_3$:

$$\begin{aligned}
L(p) = &-\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T - \lambda_0 \left( \int_{0}^{\infty} p(S_T) \, dS_T - 1 \right) \\
&- \lambda_1 \left( \int_{0}^{\infty} S_T p(S_T) \, dS_T - 102 \right) - \lambda_2 \left( \int_{0}^{\infty} S_T^2 p(S_T) \, dS_T - 10,600 \right) \\
&- \lambda_3 \left( \int_{0}^{\infty} \ln(S_T) p(S_T) \, dS_T - 4.60 \right)
\end{aligned}$$

Taking the functional derivative with respect to $p(S_T)$ and setting it to $0$:
$$\frac{\delta L}{\delta p(S_T)} = -1 - \ln p(S_T) - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T) = 0$$

Solving explicitly for $p(S_T)$:
$$\ln p(S_T) = -(1 + \lambda_0) - \lambda_1 S_T - \lambda_2 S_T^2 - \ln(S_T^{\lambda_3})$$
$$p(S_T) = \frac{1}{Z} S_T^{-\lambda_3} \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right)$$

where $Z = e^{1 + \lambda_0}$ is the normalization constant (partition function) ensuring $\int_{0}^{\infty} p(S_T) \, dS_T = 1$.

---

#### --- Interpretation of the Resulting Distribution

The derived functional form can be recognized as a generalized parameterization belonging to the **univariate exponential family**:

$$p(S_T) = \frac{1}{Z} S_T^{-\lambda_3} \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right), \quad \text{for } S_T > 0$$

* If $\lambda_2 = 0$ and $\lambda_3 = 0$, it reduces to a standard **Exponential distribution** (typical for pure positive drift without volatility tracking).
* If $\lambda_1 = 0$ and $\lambda_2 > 0$, it relates to a one-sided **Log-Gamma / truncated Gaussian alternative** depending on the power coefficient.
* With all three multipliers active ($\lambda_1, \lambda_2, \lambda_3$), it establishes a highly flexible, non-Gaussian profile capable of accommodating the heavy downside skew and non-linear penalty terms introduced by out-of-the-money put options.
