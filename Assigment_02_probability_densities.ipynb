#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



1. Construct a probability space  (Ω,F,P)  suitable for this model.

To model this experiment formally, we define the three components of the probability space:

Sample Space ($\Omega$): The set of all possible outcomes for the number of failures.$\Omega = \{0, 1, 2, 3\}$


$\sigma$-algebra ($\mathcal{F}$): Since the sample space is finite, we use the power set (all possible subsets of $\Omega$).

$\mathcal{F} = 2^\Omega$Probability Measure ($P$): The function that assigns probabilities to the events in $\mathcal{F}$ based on the given values:$P(\{0\}) = 0.50, P(\{1\}) = 0.30, P(\{2\}) = 0.15, P(\{3\}) = 0.05$

2. Define  X  as a random variable on this space.

A random variable $X$ is a function $X: \Omega \to \mathbb{R}$. For this model, $X$ is defined as the identity function:$$X(\omega) = \omega, \quad \text{for all } \omega \in \Omega$$This means if the outcome is 2 failures, the random variable $X$ takes the value 2.

3. Determine the probability mass function of  X .

The PMF $p_X(x) = P(X = x)$ is given by:$$p_X(x) = \begin{cases} 0.50 & \text{if } x = 0 \\ 0.30 & \text{if } x = 1 \\ 0.15 & \text{if } x = 2 \\ 0.05 & \text{if } x = 3 \\ 0 & \text{otherwise} \end{cases}$$

4. Compute the expectation  E[X] .

$$E[X] = \sum_{i} x_i P(X = x_i)$$$$E[X] = (0 \times 0.50) + (1 \times 0.30) + (2 \times 0.15) + (3 \times 0.05) = \mathbf{0.75}$$

5. Compute the variance  Var(X) .

First, calculate $E[X^2]$:$$E[X^2] = (0^2 \times 0.50) + (1^2 \times 0.30) + (2^2 \times 0.15) + (3^2 \times 0.05) = 1.35$$Now, use the variance formula:$$Var(X) = E[X^2] - (E[X])^2 = 1.35 - (0.75)^2 = 1.35 - 0.5625 = \mathbf{0.7875}$$

6. Find  P(X≥1)  and  P(X≥2) .

$P(X \ge 1)$: $P(X=1) + P(X=2) + P(X=3) = 0.30 + 0.15 + 0.05 = \mathbf{0.50}$$P(X \ge 2)$: $P(X=2) + P(X=3) = 0.15 + 0.05 = \mathbf{0.20}$

7. Briefly explain the practical meaning of the expectation and variance in this setting.

Expectation ($0.75$): This is the long-term average. If we record failures over many days, we expect an average of 0.75 failures per day. It helps the district estimate daily workload.

Variance ($0.7875$): This represents the spread or volatility. It tells the district how much the actual daily failures typically deviate from the average, which is crucial for planning "worst-case" staffing needs.

4. Verify that  f  is a valid probability density function.

A valid PDF must satisfy two conditions:Non-negativity: Since $x \in (0, 1)$, then $x > 0$ and $(1-x) > 0$. The Beta function $B(2,18)$ is also positive. Thus, $f(x) \ge 0$.Total area equals 1:$$\int_{0}^{1} \frac{1}{B(2, 18)} x^{2-1} (1-x)^{18-1} \, dx = \frac{1}{B(2, 18)} \int_{0}^{1} x^{2-1} (1-x)^{18-1} \, dx$$By the definition of the Beta function, the integral equals $B(2, 18)$.

Therefore, $\frac{B(2, 18)}{B(2, 18)} = 1$.

 For a Beta distribution, $E[X] = \frac{\alpha}{\alpha + \beta}$:$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$

The support of a continuous random variable is the set of all possible values it can take.$$Support: X \in [4, 10]$$

2. Write down the probability density function  fX(x) .

For a Triangular distribution with $a=4, b=10, c=6$:$$f_X(x) = \begin{cases} \frac{2(x-4)}{(10-4)(6-4)} = \frac{x-4}{6} & \text{for } 4 \le x < 6 \\ \frac{2(10-x)}{(10-4)(10-6)} = \frac{10-x}{12} & \text{for } 6 \le x \le 10 \\ 0 & \text{otherwise} \end{cases}$$

We integrate over the two pieces of the piecewise function:$$\int_{4}^{6} \frac{x-4}{6} dx + \int_{6}^{10} \frac{10-x}{12} dx$$First part: $\frac{1}{6} [\frac{x^2}{2} - 4x]_4^6 = \frac{1}{6} [(18-24) - (8-16)] = \frac{1}{6}[-6 - (-8)] = \frac{2}{6} = \frac{1}{3}$

Second part: $\frac{1}{12} [10x - \frac{x^2}{2}]_6^{10} = \frac{1}{12} [(100-50) - (60-18)] = \frac{1}{12}[50 - 42] = \frac{8}{12} = \frac{2}{3}$$\frac{1}{3} + \frac{2}{3} = 1$$

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
